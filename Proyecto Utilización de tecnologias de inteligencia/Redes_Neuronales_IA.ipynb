```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.gaussian_process.kernels import RBF
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

# Paso 2: cargar datos
iris = datasets.load_iris()
X = iris.data[:, :2]  # tomamos solo las dos primeras características
y = iris.target

# Seleccionar solo dos clases para visualización binaria
X = X[y != 2]
y = y[y != 2]

# Paso 3: dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Paso 4: entrenar modelo SVM lineal
model = SVC(kernel='linear', C=1)
model.fit(X_train, y_train)

# Paso 5: evaluar modelo
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# Paso 6: visualizar el margen y los vectores soporte
def plot_svm(model, X, y):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='autumn')

    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # crear malla
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = model.decision_function(xy).reshape(XX.shape)

    # dibujar márgenes y frontera
    ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.7, linestyles=['--', '-', '--'])

    # vectores soporte
    ax.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
               s=100, linewidth=1, facecolors='none', edgecolors='k')

    plt.title("SVM con kernel lineal")
    plt.show()

plot_svm(model, X, y)
```

```python
import numpy as np

# Función objetivo simple: f(x) = x^2 (mínimo en x=0)
def objective_function(x):
    return x**2

# Búsqueda local: descenso del gradiente simple
def local_search(x):
    alpha = 0.1  # tasa de aprendizaje
    return x - alpha * (2 * x)  # derivada de x^2 = 2x

# Inicializar población
np.random.seed(42)
population = np.random.uniform(-10, 10, size=10)

# Configuración del algoritmo
generations = 20
mutation_rate = 0.2

for gen in range(generations):
    fitness = [objective_function(ind) for ind in population]

    sorted_indices = np.argsort(fitness)
    population = population[sorted_indices]  # ordenar por aptitud

    # Cruza: promedio de los mejores dos
    child = (population[0] + population[1]) / 2

    # Mutación aleatoria
    if np.random.rand() < mutation_rate:
        child += np.random.normal(0, 1)

    # Búsqueda local
    child = local_search(child)

    # Reemplazo del peor individuo
    population[-1] = child

    print(f"Gen {gen}: mejor solución = {population[0]:.4f}, f(x) = {objective_function(population[0]):.4f}")
```

```python
import numpy as np

# Función objetivo simple: f(x) = x^2 (mínimo en x=0)
def objective_function(x):
    return np.sum(x**2)

# Búsqueda local: descenso del gradiente simple
def local_search(x):
    alpha = 0.1  # tasa de aprendizaje
    return x - alpha * (2 * x)  # derivada de x^2 = 2x

dimension = 5  # Dimensión del problema
# Inicializar población
np.random.seed(42)
population = np.random.uniform(-10, 10, size=(10, dimension))

# Configuración del algoritmo
generations = 20
mutation_rate = 0.2

for gen in range(generations):
    fitness = [objective_function(ind) for ind in population]

    sorted_indices = np.argsort(fitness)
    population = population[sorted_indices]  # ordenar por aptitud

    # Cruza: promedio de los mejores dos
    child = (population[0] + population[1]) / 2

    # Mutación aleatoria
    if np.random.rand() < mutation_rate:
        child += np.random.normal(0, 1)

    # Búsqueda local
    child = local_search(child)

    # Reemplazo del peor individuo
    population[-1] = child

    mejor = population[0]
    print(f"Gen {gen}: mejor solución = {np.round(mejor, 4)}, f(x) = {objective_function(mejor):.4f}")
```

```python
from minisom import MiniSom
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

# Cargar y normalizar datos
data = load_diabetes()
X = data.data
y = data.target

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Crear SOM: 7x7 mapa, 4 características (Iris)
som = MiniSom(x=15, y=15, input_len=X.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(X_scaled)
som.train_random(X_scaled, num_iteration=100)



# Contar cuántos caen en cada celda
cell_counts = defaultdict(int)

for x in X_scaled:
    w = som.winner(x)
    cell_counts[w] += 1

# Dibujar celdas con conteo
plt.figure(figsize=(8, 7))
for position, count in cell_counts.items():
    plt.text(position[0] + 0.5, position[1] + 0.5, str(count),
             color='red', ha='center', va='center', fontsize=10, weight='bold')

plt.title("Conteo de datos por celda (SOM - Diabetes)")
plt.grid()
plt.xlim(0, 15)
plt.ylim(0, 15)
plt.show()

```

Resultados con K-means

```python

from sklearn.cluster import KMeans



# ----------- K-means (versión tipo mapa) -----------

# K significa número de clusters; usamos 15*15 para simular la misma cantidad de celdas que el SOM (opcional)
kmeans = KMeans(n_clusters=15, random_state=42)
labels = kmeans.fit_predict(X_scaled)

# Simular posiciones de los clusters en una grilla 2D para visualización (sin topología real)
positions = {}
for i, label in enumerate(np.unique(labels)):
    # asignar coordenadas (0,0), (1,0), ..., (4,2), etc.
    x = i % 5
    y = i // 5
    positions[label] = (x, y)

# Contar cuántos puntos hay en cada clúster
cluster_counts = defaultdict(int)
for label in labels:
    cluster_counts[label] += 1

# Visualizar conteo tipo mapa
plt.figure(figsize=(8, 7))
for label, count in cluster_counts.items():
    pos = positions[label]
    plt.text(pos[0] + 0.5, pos[1] + 0.5, str(count),
             color='blue', ha='center', va='center', fontsize=10, weight='bold')

plt.title("Conteo por clúster (K-means - Diabetes, visualización tipo mapa)")
plt.grid()
plt.xlim(0, 5)
plt.ylim(0, (len(cluster_counts) // 5) + 1)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from minisom import MiniSom
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Cargar datos
data = load_diabetes()
X = data.data
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# --- K-MEANS ---
kmeans = KMeans(n_clusters=15, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# --- SOM ---
som = MiniSom(x=15, y=15, input_len=X.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(X_scaled)
som.train_random(X_scaled, 100)

# Convertir posiciones (x, y) del ganador a un único número de clúster
som_labels = [som.winner(x)[0]*15 + som.winner(x)[1] for x in X_scaled]

# --- MÉTRICAS ---
def evaluar(nombre, etiquetas):
    sil = silhouette_score(X_scaled, etiquetas)
    db = davies_bouldin_score(X_scaled, etiquetas)
    ch = calinski_harabasz_score(X_scaled, etiquetas)
    print(f"\n{nombre} clustering:")
    print(f"  Cohesión y separación de clústeres: {sil:.4f}")
    print(f"  Indice Separación entre clústeres: {db:.4f}")
    print(f"  Indice Varianza entre vs. dentro de clústeres: {ch:.2f}")

evaluar("K-means", kmeans_labels)
evaluar("SOM", som_labels)
```

```python
from minisom import MiniSom
from sklearn.datasets import load_diabetes
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict

# Cargar y normalizar datos
data = load_diabetes()
X = data.data
y = data.target

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Crear SOM: 7x7 mapa, 4 características (Iris)
som = MiniSom(x=15, y=15, input_len=X.shape[1], sigma=1.0, learning_rate=0.5)
som.random_weights_init(X_scaled)
som.train_random(X_scaled, num_iteration=100)



# Crear U-Matrix (distancia media a vecinos directos)
weights = som.get_weights()
u_matrix = np.zeros((15, 15))

# Función para obtener vecinos válidos (dentro del mapa)
def get_neighbors(i, j, max_x, max_y):
    neighbors = []
    for dx, dy in [(-1,0), (1,0), (0,-1), (0,1)]:  # arriba, abajo, izq, der
        ni, nj = i + dx, j + dy
        if 0 <= ni < max_x and 0 <= nj < max_y:
            neighbors.append((ni, nj))
    return neighbors

# Calcular distancia promedio a vecinos
for i in range(15):
    for j in range(15):
        w_ij = weights[i, j]
        neighbors = get_neighbors(i, j, 15, 15)
        distances = [np.linalg.norm(w_ij - weights[ni, nj]) for ni, nj in neighbors]
        u_matrix[i, j] = np.mean(distances)

# Mostrar la U-Matrix
plt.figure(figsize=(8, 7))
plt.imshow(u_matrix.T, cmap='bone_r', origin='lower')
plt.colorbar(label='Distancia media a vecinos')
plt.title("U-Matrix (Mapa de distancias del SOM - Diabetes)")
plt.grid(False)
plt.show()
```

Pytorch

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from torch.utils.data import DataLoader, TensorDataset

# Datos
digits = load_digits()
X = digits.data
y = digits.target

# Escalado
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Conversión a tensores
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train_tensor = torch.FloatTensor(X_train)
X_test_tensor = torch.FloatTensor(X_test)
y_train_tensor = torch.LongTensor(y_train)
y_test_tensor = torch.LongTensor(y_test)

# Dataset y DataLoader
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Modelo MLP
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(64, 100)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(100, 10)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = MLP()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Entrenamiento
for epoch in range(30):
    for inputs, labels in train_loader:
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

# Evaluación
with torch.no_grad():
    outputs = model(X_test_tensor)
    _, y_pred = torch.max(outputs, 1)

print(classification_report(y_test_tensor, y_pred))
```

Construye una red neuronal multicapa con Keras o PyTorch para clasificar el dataset Fashion MNIST. Evalúa con precisión y matriz de confusión.



```python
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar datos
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Normalizar y convertir etiquetas a one-hot
X_train, X_test = X_train / 255.0, X_test / 255.0
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

# Construir modelo
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compilar y entrenar
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train_cat, epochs=10, batch_size=32, validation_split=0.1, verbose=1)

# Evaluar
loss, acc = model.evaluate(X_test, y_test_cat)
print(f"\nPrecisión en test: {acc:.4f}")

# Matriz de confusión
y_pred = model.predict(X_test).argmax(axis=1)
cm = confusion_matrix(y_test, y_pred)
print("Matriz de Confusión:\n", cm)
print(classification_report(y_test, y_pred))
```
